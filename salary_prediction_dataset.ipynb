# Salary Prediction DataSet

```python
import pandas as pd
```

```python
df=pd.read_csv("salary_prediction_dataswt.csv")
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df[['Experience','Projects']],df[['Salary']],test_size=0.3,random_state=42)
```

```python
reg=linear_model.LinearRegression()
```
