```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df =pd.read_csv("homeprice.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,marker='+')
# plt.plot(df.area,reg.predict(area),color='red')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python

reg.predict([[5000]])
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='red')
```

# save and load trained model using:
1- pickle
2-sklearn

```python
import pickle
```

```python
with open('model_pickle','wb') as f:
    pickle.dump(reg,f)
```

```python
with open('model_pickle','rb') as f:
    model=pickle.load(f)
```

```python
model.predict([[5000]])
```

# joblib also save and load model in same way ,but joblib is more efficient on object that carry large  numpy arrays

```python
import joblib
```

```python
joblib.dump(reg,'model_joblib')
```

```python
mj=joblib.load('model_joblib')
```

```python
mj.predict([[5000]])
```
